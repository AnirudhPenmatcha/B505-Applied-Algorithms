```python
import matplotlib.pyplot as plt
import random
```

```python
collision_index = []
for _ in range(1000):
  birthday_set = set()
  birthday = random.randint(1, 366)


  while(birthday not in birthday_set):
    birthday_set.add(birthday)
    birthday = random.randint(1, 366)

  collision_index.append(len(birthday_set))
  print(collision_index)
```

```python
plt.hist(collision_index, bins=15)
plt.show()
```

```python
# Question 1

from copy import deepcopy
import math

def helpRyan(points, sniperPoint):
    def simplify(points, sniper):
        simplified = []
        q = sniper
        for i in range(len(points)):
            p = [points[i][0], points [i][1]]
            a = math.dist(p, q); b = math.sqrt(points[i][2]); c = i
            simplified.append([a+b, c])
        return simplified
        
    def mergeSort(simplified):
        if len(simplified) > 1:
            mid = len(simplified)//2
            L = simplified[:mid]
            R = simplified[mid:]
            mergeSort(L)
            mergeSort(R)
            
            i = j = k = 0
 
            while i < len(L) and j < len(R):
                if L[i][0] <= R[j][0]:
                    simplified[k] = L[i]
                    i += 1
                else:
                    simplified[k] = R[j]
                    j += 1
                k += 1

            while i < len(L):
                simplified[k] = L[i]
                i += 1
                k += 1

            while j < len(R):
                simplified[k] = R[j]
                j += 1
                k += 1
        return simplified
    

    points = deepcopy(points); sniperPoint = deepcopy(sniperPoint)
    simplified = simplify(points, sniperPoint)
    simplified = mergeSort(simplified)
    output = [points[simplified[j][1]] for j in range(len(simplified))]
    
    return output
```

```python
# Question 2

def divide(arr, low, high):
    pivot_val = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot_val:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_select(arr, low, high, k):
    while low <= high:
        pivot_index = divide(arr, low, high)
        if pivot_index == k:
            return arr[pivot_index]
        elif pivot_index < k:
            low = pivot_index + 1
        else:
            high = pivot_index - 1
    return None

def jacksMachine(objects, query) -> int:
    if query > len(objects):
        return -1
    return quick_select(objects, 0, len(objects) - 1, len(objects) - query)
```

```python
# Question 4

def find_last_non_conflicting(events, n):
    low, high = 0, n - 1
    while low <= high:
        mid = (low + high) // 2
        if events[mid][1] <= events[n][0]:
            if events[mid + 1][1] <= events[n][0]:
                low = mid + 1
            else:
                return mid
        else:
            high = mid - 1
    return -1

def maxEarnings(start_time, end_time, profit):
    events = sorted(zip(start_time, end_time, profit), key=lambda x: x[1])
    dp = [0 for _ in range(len(events))]
    dp[0] = events[0][2]

    for i in range(1, len(events)):
        incl_prof = events[i][2]
        l = find_last_non_conflicting(events, i)
        if l != -1:
            incl_prof += dp[l]
        dp[i] = max(incl_prof, dp[i - 1])

    return dp[-1]

start_time = [1, 2, 3, 3]
end_time = [3, 4, 5, 6]
profit = [50, 10, 40, 70]

print(maxEarnings(start_time, end_time, profit))
```

```python
# Question 6

def maxCoins(nums):
    nums = [1] + nums + [1]
    length = len(nums)
    dp = [[0] * length for _ in range(length)]
    for balloon_length in range(2, length):
        for left_index in range(0, length - balloon_length):
            right_index = left_index + balloon_length
            for k_index in range(left_index + 1, right_index):
                dp[left_index][right_index] = max(dp[left_index][right_index], nums[left_index] * nums[k_index] * nums[right_index] + dp[left_index][k_index] + dp[k_index][right_index])
    return dp[0][length - 1]
```

```python
# Question 7

from heapq import heapify, heappush, heappop 
 
    
class Huffman():
    def __init__(self):
        self.huffman_codes = {}
        self.source_string = ""
        
    def set_source_string(self, src_str):
        self.source_string = src_str
        
    def generate_codes(self):
        huffman_codes = {}
        # Get the letter count 
        for letter in self.source_string:
            if letter in huffman_codes.keys():
                huffman_codes[letter] += 1
            else:
                huffman_codes[letter] = 1
        
        # Create a min-heap from the frequency counts
        heap = [(freq, [char]) for char, freq in huffman_codes.items()]
        heapify(heap)
        
        # Step 2 and 3
        while len(heap) > 1:
            # Extract the two minimum elements from the heap
            freq1, chars1 = heappop(heap)
            freq2, chars2 = heappop(heap)
            
            # Assign codewords
            for char in chars1:
                self.huffman_codes[char] = '0' + self.huffman_codes.get(char, '')
            for char in chars2:
                self.huffman_codes[char] = '1' + self.huffman_codes.get(char, '')
            
            # Merge the two least frequent symbols and update the heap
            heappush(heap, (freq1 + freq2, chars1 + chars2))
        
        
        # Reverse all codewords generated
#         print(self.huffman_codes)
        sorted(self.huffman_codes, reverse=True)
        print(self.huffman_codes)
        
    def encode_message(self, message_to_encode):
        encoded_msg = ""
        # Write you code here
        for char in message_to_encode:
            encoded_msg += self.huffman_codes[char]
        
        return encoded_msg
    
    def decode_message(self, encoded_msg):
        decoded_msg = ""
        current_code = ""
        for bit in encoded_msg:
            current_code += bit
            for char, code in self.huffman_codes.items():
                if code == current_code:
                    decoded_msg += char
                    current_code = ""
                    break

        return decoded_msg
        
```
